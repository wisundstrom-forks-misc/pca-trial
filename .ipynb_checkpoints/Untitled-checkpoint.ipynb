```python
df= pd.read_csv("data/ysnp.csv")
```

```python
keys=list(df.keys())
keys
```

```python

features=keys[0:11]+keys[14:]
df2=df.loc[:, features]
```

```python
null_columns=df2.columns[df2.isnull().any()]
df2[null_columns].isnull().sum()
```

```python
badrows=df2[df2.isnull().any(axis=1)][null_columns]
df2=df2.drop(badrows.index)
y=df2['Recreation Visits']
features=features[1:]
```

```python

x = df2.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

#y is defined above
ax.scatter(principalDf['principal component 1'], principalDf['principal component 2'], c=y, cmap=plt.cm.coolwarm)
xvector = pca.components_[0]
yvector = pca.components_[1]
score=principalComponents[:,0:2]
xs = score[:,0]
ys = score[:,1]

for i in range(len(xvector)):
# arrows project features (ie columns from csv) as vectors onto PC axes
    plt.arrow(0, 0, xvector[i]*max(xs), yvector[i]*max(ys), color='r', width=0.0005, head_width=0.0025)
    plt.text(xvector[i]*max(xs)*1.1, yvector[i]*max(ys)*1.1, features[i], color='r')

ax.grid()
```
